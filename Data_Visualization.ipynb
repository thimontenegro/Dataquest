# Múltiplos Gráficos

Quando trabalhamos com um único gráfico, o pyplot armazena e vai atualizando este único estado.
Quando estamos trabalhando com múltiplos gráficos devemos especificar ao pyplot qual gráfico queremos modificar. 

Mas para especificar qual o gráfico queremos mexer, vamos entender como o pyplot gera um único gráfico:



*   O pyplot gera um container para todos os gráficos criados (Ele retorna um Objeto Chamado **Figure**)
*   Este container são posicionados no grid (Retorna um objeto **Axes**)
*   Em seguida são adicionado os símbolos visuais - como linhas, pontos etc. -

O objeto figure funciona como um container, armazena, todos os nossos gráficos para customizar a aparencia e o comportamento dos gráficos.

Nos podemos criar manualmente o objeto Figure através:

```
# fig = plt.figure()
```

Ao invés de chamar somente pela função no pyplot, vamos salvar o retorno desse objeto em uma variável chamada fig.

Depois da variável fig é criada, são criados eixos,*O objeto axes ali em cima*, para um único gráfico que não possui nenhum dado.

O objeto axes atua como seu próprio conteiner para os componentes para exibir o gráfico, como:



*   Valores nos eixos X e Y;
*   Símbolos visuais como, linhas e grids.

Enquanto os gráficos são representados usando instances do Objeto *Axes*, são referidos como subplots no matplotlib.

Para adicionar um novo subplot em uma *Figure* já criada, nós usamos:



```
# eixo_obj = fig.add_subplot(nrows, ncols, plot_number)
```
Se queremos uma figura que possua 4 gráficos, precisamos escrever.

```
eixo_1 = fig.add_subplot(2,1,1)
eixo_2 = fig.add_subplot(2,1,2)
eixo_3 = fig.add_subplot(2,1,3)
eixo_4 = fig.add_subplot(2,1,4)
``` 















````
fig = plt.figure()
eixo_1 = fig.add_subplot(2,1,1)
eixo_2 = fig.add_subplot(2,1,2)
plt.show()
````

Para cada subplot, o matplotlib gera uma coordenada no grid.

Quando criamos um subplot, o matplotlib reconhece e cria um grid com 2 fileiras e 2 colunas. Ao adicionar cada subplot, precisamos especificar valores que correspondem ao objeto *Axes*.

No matplotlib, o número é inicializado na posição localizada mais a esquerda (top left) e vai se organizando de acordo com o restante das posições nas fileiras.

Vamos ver alguns exemplos:


````
fig = plt.figure()
eixo_1 = fig.add_subplot(4,1,1)
eixo_2 = fig.add_subplot(4,1,2)
eixo_3 = fig.add_subplot(4,1,3)
eixo_4 = fig.add_subplot(4,1,4)
plt.show()
````

````
fig = plt.figure()
eixo_1 = fig.add_subplot(2,2,1)
eixo_2 = fig.add_subplot(2,2,2)
eixo_3 = fig.add_subplot(2,2,3)
eixo_4 = fig.add_subplot(2,2,4)
plt.show()
````

````
fig = plt.figure()
eixo_1 = fig.add_subplot(2,3,1)
eixo_2 = fig.add_subplot(2,3,2)
eixo_3 = fig.add_subplot(2,3,3)
eixo_4 = fig.add_subplot(2,3,4)
plt.show()
````

Para gerar o gráfico em linhas através dos subplots, precisamos chamar o método 

```
Axes.plot(x,y)
```



````
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(unrate[0:10]['DATE'], unrate[0:10]['VALUE'])
ax2.plot(unrate[10:20]['DATE'],unrate[10:20]['VALUE'])
plt.show()
````

Um problema recorrente quando desejamos plotar 2 gráficos, é que os eixos X acabam muitas vezes ilegiveis.
Outro fator é que os gráficos ficam muito apertados o que gera um problema para interpreta-los.

Isso se torna difícil por que a biblioteca utiliza de valores padrões para a área total para plotar os gráficos. Se queremos alterar esses valores precisamos especificar quais as dimensões da nossa nova área.

Para isso usaremos a parametro do método figure(), chamada figsize.

Esse parametro possui uma tupla de números flutuantes:



```
fig = plt.figure(figsize = (largura, altura))
```



````
fig = plt.figure(figsize = (12,8))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(unrate[0:10]['DATE'],unrate[0:10]['VALUE'])
ax1.set_title('Taxa de Desemprego Mensal em 1948')
ax2.plot(unrate[10:20]['DATE'],unrate[10:20]['VALUE'])
ax2.set_title('Taxa de Desemprego Mensal em 1949')
plt.show()
````

Contudo.....
Dividir gráficos em subplots pode ser uma boa alternativa se você deseja mostrar diferentes situações.
Contudo se desejamos realizar comparações entre os dois gráficos pode ser uma tarefa um pouco mais complicada.

Para isso talvez uma solução mais adequada seja plotar diferentes funções afim de realizar a comparação no mesmo gráfico.

Para isso basta realizar mais de uma chamada do plt.plot(x,y).

````
unrate['MONTH'] = unrate['DATE'].dt.month

fig = plt.figure(figsize = (6,6))
plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'], c = 'blue')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['VALUE'], c = 'red')
plt.title('Comparativo entre a Taxa de Desemprego entre 1948 e 1949')
plt.show()
````

Se desejamos trabalhar com mais do que 2 subplots, definir isso pode ser bem cansativos. 
Ao invés de definir manualmente podemos criar uma laço for e a cada iteração gerar um novo subplot alterando seus valores.

````
fig = plt.figure(figsize = (10,8))

for i in range(4):
  ax = fig.add_subplot(4,1,i+1)
  index_inicio = i * 12
  index_fim = (i + 1) * 12
  subset = unrate[index_inicio: index_fim]
  ax.plot(subset['DATE'],subset['VALUE'])

plt.show()
  
````

Ou se desejamaos trabalhar com multiplos gráficos linhas em um único plot.

````
fig = plt.figure(figsize = (10,6))
colors = ['red', 'blue', 'green', 'darkorchid', 'black']
for i in range(5):
    start_index = i * 12
    end_index = (i + 1) * 12
    subset = unrate[start_index:end_index]
    plt.plot(subset['MONTH'], subset['VALUE'], c = colors[i])
plt.show()
````

Quando generamos múltiplos gráficos em linha, precisamos especificar qual cor está ligada a quê. Para isso vamos usar o parametro da classe pyplot.plot() chamado label, que nós ajuda a identificar.

# Gráfico de Barra

O gráfico de barra (barplot/barchart)  é o tipo mais comum de gráfico. Onde mostra a realão entre uma variável númerica e uma variável categórica.
Gráficos de barra são frequentemente confudidos com histogramas, que são coisas completamente distintas!

Histogramas são gráficos na quais possui apenas uma única variável númerica!


Vamos utilizar um dataset chamado **Fandango Score**
que é basicamente um site que agrega o contéudo de ratings de filmes de diversos sites, como O Rotten Tomatoes, Metacrits e IMBD.

Para ter uma ídeia o dataset é composto por:


*   FILM - Nome do filme
*   RT_user_norm - A média das avaliações do Rotten Tomatoes, em uma escala de 1 a 5 

*   Metacritic_user_nom - A média das avaliações do Metacriitic, em uma escala de 1 a 5
*   IMDB_norm - A média das avaliações do IMDB, em uma escala de 1 a 5
*  Fandango_Ratingvalue -  A média das avaliações do Fandango, em uma escala de 1 a 5
* Fandango_Stars - O ranking de avaliação no site do fandango.



````
import pandas as pd
df = pd.read_csv('fandango_scores.csv')
cols = ['FILM', 'RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
norm_reviews = df[cols]
norm_reviews.head()
````

Quando criamos um gráfico em linha, o plt.plot() bastava passar os dados e ele fazia o resto, certo?

Contudo para criarmos um gráfico em barra, precisamos especificar as posições no gráfico, a largura das barras e a posição nos eixos.

Para gerar um gráfico em barra vertical, precisamos utilizar o plt.bar() ou Axes.bar().

Vamos utilizaro Axes.bar() pelo simples detalhe que podemos customizar nossos gráficos mais facilmente.

Também vamos utilizar o plt.subplots() para gerar um único subplot, gerando um objeto **Figure** e outro **Axes**.



```
# fig, ax = plt.subplots()
```

O método Axes.bar() precisa de dois parâmetros, left e height.

Usamos o parâmetro Left para especificar as coordenadas no eixo x.

O parâmetro height, utilizamos para especificar a altura de cada barra.

*PS: Os dois parâmetros acima aceitam listas!*

A função np.arange() retorna um espaçamento igual de valores.



````
import matplotlib.pyplot as plt
from numpy import arange
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

bar_heights = norm_reviews[num_cols].iloc[0].values
bar_positions = arange(5) + 0.75

fig, ax = plt.subplots()
ax.bar(bar_positions, bar_heights, 0.5)
plt.show()
````

Por default, o matplotlib define os eixos-x em valores númericos que se estende por todo o gráfico.
Para posicionar as barras, utilizamos o Axes.set_xticks() para alterar as posições no eixo x.

x_ticks aceita um argumento que podemos passar uma lista de objetos, números para cada posição das nossas barras.



````
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
bar_heights = norm_reviews[num_cols].iloc[0].values
bar_positions = arange(5) + 0.75
tick_positions = range(1,6)
fig, ax = plt.subplots()
ax.bar(bar_positions, bar_heights, 0.5)
ax.set_xticks(tick_positions)
ax.set_xticklabels(num_cols, rotation = 90)
plt.show()
````

Também podemos criar gráficos horizontais, para isso basta ao invés de usarmos axes.bar(), utilizar axes.barh().



```
# Axes.barh(bottom, width)

Bottom = Especificar a localização na coordenada Y.
width = é o parâmetro para especificar o tamanho das barras.

```



````
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
bar_width = norm_reviews[num_cols].iloc[0].values
bar_position = arange(5) + 0.75
tick_positions = range(1,6)
fig, ax = plt.subplots()
ax.barh(bar_positions, bar_width, 0.5)
ax.set_yticks(tick_positions)
ax.set_yticklabels(num_cols)
plt.show()
````

# Gráficos de Dispersão
Um gráfico de dispersão exibe o valor de 2 conjuntos de dados em 2 dimensões. Cada ponto representa uma observação. A posição no eixo X (horizontal) e Y (vertical) representa os valores das duas variáveis.
É de grande ajuda para enteder a correlação entre duas variáveis.
É comum utilizar outras features para demonstrar a relação entre essas duas variáveis, como cor e tamanho.


Para gerar o gráfico de dispersão, temos duas posibilidades.
Utilizar o plt.scatter(x,y) ou axes.scatter(x,y).




````
plt.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['RT_user_norm'])
plt.xlabel('Fandango')
plt.ylabel('Rotten Tomatoes')
plt.show()
````

O gráfico de dispersão sugere se ocorre uma fraca ou forte correlação.
Por exemplo no gráfico acima, a correlação entre as avaliações do Fandango e do Rotten Tomatoes é positiva. Pois a medida que que o eixo X cresce o eixo Y também cresce.

No Gráfico de dispersão para entender se duas variávies estão correlacionadas, não importa qual está no eixo x ou no eixo y. Isso corre porque o gráfico captura e exibe a correlação da mesma forma.

Vamos gerar um gráfico apenas alterando os eixos.


````
fig = plt.figure(figsize=(5,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['RT_user_norm'])
ax1.set_xlabel('Fandango')
ax1.set_ylabel('Rotten Tomatoes')
ax2.scatter(norm_reviews['RT_user_norm'], norm_reviews['Fandango_Ratingvalue'])
ax2.set_xlabel('Rotten Tomatoes')
ax2.set_ylabel('Fandango')
plt.show()
````

O segundo gráfico é apenas o espelho da primeira, e a natureza da corelação é mantida!


Se queremos gerar múltiplos gráficos de dispersão afim de compara-los. É importante que todos os gráficos compartilhem os mesmos eixos x e y. Para que os alcances dos gráficos sejam semelhantes. O que no gráfico acima não existe.

Para isso podemos utilizar o:



```
# ax.set_xlim(inicio, fim)
  ax.set_ylim(inicio, fim)
```




````
fig = plt.figure(figsize = (5,10))
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
ax3 = fig.add_subplot(3,1,3)
ax1.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['RT_user_norm'])
ax1.set_xlabel('Fandango')
ax2.set_ylabel('Rotten Tomatoes')
ax1.set_xlim(0,5)
ax2.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['Metacritic_user_nom'])
ax2.set_xlabel('Fandango')
ax2.set_ylabel('Metacritic')
ax2.set_xlim(0,5)
ax3.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['IMDB_norm'])
ax3.set_xlabel('Fandango')
ax3.set_ylabel('IMDB')
ax3.set_xlim(0,5)
plt.show()
````

# Histogramas
O Histograma pega a entrada de uma única varíavel númerica. Está variável é dividida entre intervalos idênticos, chamados de bins. 
O número de observações por intervalo é representado pela altura da barra.
Vale notar que a forma de cada histograma pode ser diferente de acordo com o número de bins que você utiliza, para isso utilize diversos valores de bins antes de tomar qualquer conclusão.


De ínicio vamos realizar a distribuição de frequência das avaliações do fandago e IMDB.

A distribuição de frequência consite de valores únicos em uma coluna acompanhado com o total de valores da sua respectiva coluna.

Utilizamos value_counts() em um objeto Series retornando assim a distribuição de frequência daquela coluna.


````
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('fandango_scores.csv')
````

````
freq_count = df['Fandango_Ratingvalue'].value_counts()
freq_count
````

Podemos ordenar os valores, utilizando após o value_counts o sort_index().

````
freq_count = df['Fandango_Ratingvalue'].value_counts().sort_index()
freq_count

````

Ordenando de forma crescente, basta adicionar o parâmetro ascending = False no sort_index().


````
freq_count = df['Fandango_Ratingvalue'].value_counts().sort_index(ascending = False)
freq_count
````

Podemos gerar um histograma usando  tanto o plt.hist() ou axes.hist(). Este método possui apenas 1 parâmetro necessário, um objeto iterável que contém os valores para os quais queremos um histograma. Por padrão, o matplot gera:



*   Calcular o valor mínimo e máximo a partir da sequência de valores que passamos
*   Cria 10 posições no mesmo tamanho que abrangem o intervalo do valor mínimo ao máximo
*   Agrupa os valores nos bins.
*   Soma da frequência de cada bin.

O compartamento padrão do hist(), é um tanto problematico para comparar distribuições de diversas colunas.
Isso ocorre devido a divisão de cada bin, depende dos valores mínimos e máximos, ao invés de dividir a mesma estrátegia padrão de cada bin.
Para solucionar isso, utilizamos a função de range(inicio, fim).


````
plt.hist(df['Fandango_Ratingvalue'])
plt.plot()

````

````
plt.hist(df['Fandango_Ratingvalue'], range=(0,5))
plt.plot()
````

Nós podemos examinar a proporção de área que as barras estão entre 2.0 e 4.0, que determinam cerca de 50% dos filmes.
Para isso podemos aumentar o número de bins de 10 para 20 para melhorarmos a resolução da nossa distribuição.

A Nova distribuição de cada bin será de 0.25 (5/20), para adicionarmos isso, basta no segundo parâmetro do plt.hist() ou axes.hist(), informa o valor.





````
plt.hist(df['Fandango_Ratingvalue'], 20)
plt.plot()
````

# Boxplot

Um gráfico boxplot fornece um bom resumo de uma ou várias variáveis númericas.
A linha que divide o boxplot em duas partes representa a mediana dos dados. Os extremos da caixa mostra o primeiro e o terceiro quartil.
Os extremos das linhas demonstram os valores extremos (o maior valor e o menor valor) presente naquele dado.

Podemos exibir o boxplot atráves do:


```
# plt.boxplot(dado['coluna'])
```




````
plt.boxplot(df['Fandango_Ratingvalue'])
plt.show()
````

````
fig, ax = plt.subplots()
ax.boxplot(norm_reviews['Fandango_Ratingvalue'])
ax.set_ylim(0,5)
ax.set_xticklabels(['Avaliação Fandango'])
plt.show()
````

Do boxplot gerado utilizando avaliação do Fandango, podemos concluir que:


*   25% dos usúarios estão no range de 2.7 - 3.5
*   25% dos usúarios avaliaram estão entre 4.3 4.8

Para comparar diversos boxplots e comparar quais são os extremos mínimos e máximos, é preciso comparar lado a lado.

````
cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
plt.boxplot(df[cols].values)
plt.plot()
````

````
cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
fig, ax = plt.subplots()
ax.boxplot(df[cols].values)
ax.set_xticklabels(cols, rotation = 90)
ax.set_ylim(0,5)
plt.show()
````
