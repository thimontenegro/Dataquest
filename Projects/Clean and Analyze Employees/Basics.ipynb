<p>Verificando os dados
Abaixo, limpamos e exploramos as colunas cease_date e dete_start_date para garantir que todos os anos façam sentido. Usaremos os seguintes critérios:

Como cease_date é o último ano do emprego da pessoa e dete_start_date é o primeiro ano de trabalho da pessoa, não faria sentido ter anos após a data atual.
Dado que a maioria das pessoas nesse campo começa a trabalhar na casa dos 20 anos, também é improvável que o dete_start_date tenha sido antes do ano de 1940.
</p>

<p> Criando uma nova coluna</p>

```python
dete_resignations['institute_service'] = dete_resignations['cease_date'] - dete_resignations['dete_start_date']
dete_resignations['institute_service'].head()
```

<p> Identificando os índices de insatisfação</p>

```python
tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts()
```

```python
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts()
```

```python
def updated_vals(element):
    if element == '-':
        return False
    elif pd.isnull(element):
        return np.nan
    else:
        return True
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']].applymap(updated_vals).any(1, skipna = False)

tafe_resignations_up = tafe_resignations.copy()

tafe_resignations_up['dissatisfied'].value_counts(dropna = False)
```

```python
dete_resignations['dissatisfied'] = dete_resignations[['job_dissatisfaction',
       'dissatisfaction_with_the_department', 'physical_work_environment',
       'lack_of_recognition', 'lack_of_job_security', 'work_location',
       'employment_conditions', 'work_life_balance',
       'workload']].any(1, skipna=False)
dete_resignations_up = dete_resignations.copy()
dete_resignations_up['dissatisfied'].value_counts(dropna=False)
```

<h2>Combinando os datasets </h2>

```python
dete_resignations_up['institute'] = 'DETE'
tafe_resignations_up['institute'] = 'TAFE'
```

```python
combined = pd.concat([dete_resignations_up, tafe_resignations_up], ignore_index = True)

combined.notnull().sum().sort_values()
```

```python
combined_updates = combined.dropna(thresh = 500, axis = 1).copy()
combined_updates.head()
```

<h3> Limpanda as colunas</h3>

```python
combined_updates['institute_service'].value_counts(dropna = False)
```

```python

# Extract the years of service and convert the type to float
combined_updates['institute_service_up'] = combined_updates['institute_service'].astype('str').str.extract(r'(\d+)')
combined_updates['institute_service_up'] = combined_updates['institute_service_up'].astype('float')

# Check the years extracted are correct
combined_updates['institute_service_up'].value_counts()
```

```python
def transform_service(val):
    if val >= 11:
        return "Veteran"
    elif 7 <= val < 11:
        return "Established"
    elif 3 <= val < 7:
        return "Experienced"
    elif pd.isnull(val):
        return np.nan
    else:
        return "New"
combined_updates['service_cat'] = combined_updates['institute_service_up'].apply(transform_service)

# Quick check of the update
combined_updates['service_cat'].value_counts()
```

```python
combined_updates['dissatisfied'].value_counts(dropna = False)
```

```python
combined_updates['dissatisfied'] = combined_updates['dissatisfied'].fillna(False)
```
